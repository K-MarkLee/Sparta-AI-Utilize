```python
from ultralytics import YOLO
import cv2
from matplotlib import pyplot as plt
```

```python

# YOLOv8 모델 로드
model = YOLO('yolov8n.pt')  # 사전에 학습된 YOLOv8 모델을 로드

# 웹캠을 통해 실시간 객체 탐지
cap = cv2.VideoCapture(0)  # 웹캠 캡처 시작
# 웹캠 오류 처리
if not cap.isOpened():
    print("camera open failed")
    exit()
```

```python
# 웹캠 -> 

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # 프레임을 YOLOv8 모델에 입력하여 객체 탐지 수행
    # results = model(frame)

    # 카메라 창 띄우기
    results = model.predict(source = frame)
    plots = results[0].plot()
    cv2.imshow("PC_Webcam",plots)
		
		# 이후 처리코드

    # esc 누르면 종료하기
    if cv2.waitKey(1) == 27:
        break

# 웹캠 종료 및 창 닫기
cap.release()
cv2.destroyAllWindows()
```
